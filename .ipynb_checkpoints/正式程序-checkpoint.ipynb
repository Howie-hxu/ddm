```python

# coding: utf-8
# 运行说明：
#   先用鼠标单击【这里】，看到旁边的窗口指示色变成绿色，然后按下ctrl + enter
#   如果出现*说明正在运行，出现序号，表示运行结束，可以拉到最下面看运行结果
#   序号一般会递增，在没有关闭程序的情况下，反复运行序号会增加

# 每次运行前，**最好确保程序是从关闭状态启动的，直接关闭浏览器不会关闭程序**
# 关闭操作如下
# 点击左上方菜单栏的“File”，选择“Close and Halt”，此时本窗口会直接关闭
# 从DDM目录下看到，“正式程序.ipynb”的图标是灰色的，即说明程序没有运行
# 如果是绿色的，那说明其实在后台运行，没有关闭成功


import pandas as pd
import numpy as np
import media as me
import datetime,calendar,time,os


#============================================================================
## 如果日期不对，可以手动填写，比如月份不对，就直接改month，日期不对，就改day
## 因为这个是抓当前每个月的月底，如果是下一个月的月头运行，就不正确了
year = datetime.date.today().year 
month= datetime.date.today().month               # month=08

if month == 1 :
    year = year - 1
else :
    month = month - 1

day = calendar.monthrange(year, month)[1]
mmddyyyy = str(month)+"/"+str(day)+"/"+str(year)

#============================================================================

# 资料准备
originalSheet = pd.read_excel("1.xls",skiprows=7)
sameRowsDict = dict()
for index in range(len(originalSheet['科目名称'])):
    if originalSheet['科目名称'][index] not in sameRowsDict :
        sameRowsDict[originalSheet['科目名称'][index]] = list()
    sameRowsDict[originalSheet['科目名称'][index]].append(index)


# In[13]:


items,debtors,lenders,totals = list(),list(),list(),list()

for key in sameRowsDict:
    value = sameRowsDict[key]
    items.append(key)
    debtorSum = 0.0
    lenderSum = 0.0
    
    for index in value:
        debtorSum += float(originalSheet['原币借方发生额'][index])
        lenderSum +=  float(originalSheet['原币贷方发生额'][index])
    debtors.append(debtorSum)
    lenders.append(lenderSum)
    totals.append(debtorSum - lenderSum)


# In[14]:


proportionSheet = pd.read_excel("23.xls",sheetname="人员")
projectNames , projectProportions = proportionSheet["项目名称"],proportionSheet["占比"]


# In[15]:
projectProportionsList = list(projectProportions)
totalMartix = np.array(totals).reshape(len(totals),1)
projectProportionsMatrix = np.array(projectProportionsList).reshape(1,len(projectProportionsList))
result = np.dot(totalMartix,projectProportionsMatrix)
result = np.nan_to_num(result)

month = "%02d" % month
digest = str(year)[2:]+month+"上海威灵研发项目资本化分摊"

for key in me.output:
    me.output[key].clear()

markRow = list()
markColumn = list()

for i in range(len(items)):
    if(items[i] in me.selectType):
        not_all_zero = False
        for j in range(len(result[i,:])):
            if projectNames[j] in me.item2code and result[i][j] != 0:
                not_all_zero = True
                if j not in markColumn:
                    markColumn.append(j)
                nameStr = items[i]
                if items[i] in me.alianName:
                    nameStr = me.alianName[items[i]]
                me.output['预算来源'].append(me.budgetSource[items[i]]+'  '+nameStr)
                me.output['预算树'].append('2018-预算-部品事业部-先行研究中心')
                me.output['费用类型'].append(nameStr)
                me.output['借/贷类型'].append('借')
                
                if projectNames[j] == 'zongbu':
                    result[i][j] = -1.0 * result[i][j]
                    
                monkey = round(result[i][j],2)
                me.output['金额'].append(monkey)
                me.output['本位币金额'].append(monkey)
                me.output['批准金额'].append(monkey)
                
                me.output['币种'].append('人民币元')
                me.output['汇率'].append(1)
                me.output['费用发生日期'].append(mmddyyyy)
                me.output['导ERP方式'].append('GL日记账')
                me.output['摘要'].append(digest)
                me.output['项目号'].append(me.item2code[projectNames[j]])
                me.output['事业部名称'].append('部品事业部')
                me.output['businessCode'].append('30003088')
        if not_all_zero:
            markRow.append(i)
        else:
            print(items[i]+"：这个项目全部都是0")
            
for i in markRow:
    sumNum = 0.0;
    markToSum = 0;
    for j in markColumn:
        if(projectNames[j] != "zongbu"):
            markToSum = j
        pos = markRow.index(i) * len(markColumn) + markColumn.index(j)
        #print('markRow.index(',i,')  = ',markRow.index(i),'len(markColumn) = ',len(markColumn),'markColumn.index(',j,')',markColumn.index(j))
        sumNum += me.output['金额'][pos]
    rightPos = markRow.index(i) * len(markColumn) + markColumn.index(markToSum)
    if np.abs(sumNum) > 0.005:
        print("-------------------------------------------------------------\n\
            ===>>> \"%s\" 存在误差，求和-总数 = %.2f ，\n\
            \t这个数值会反应到部门 \"%s\" \n\
            \t===>值由 %.2f -> %.2f" % (items[i],sumNum,projectNames[markToSum],me.output['金额'][rightPos],(me.output['金额'][rightPos] - sumNum)))
    me.output['金额'][rightPos] = me.output['金额'][rightPos] - sumNum
    me.output['本位币金额'][rightPos] =  me.output['金额'][rightPos]
    me.output['批准金额'][rightPos] =  me.output['金额'][rightPos]


emptyList = list()

for index in range(len(me.output['预算来源'])):
    emptyList.append('')

sumSheet = pd.DataFrame()
for key in me.output:
    if(len(me.output[key]) == 0):
        me.output[key] = emptyList
    sumSheet[key] = me.output[key]

# 写入到excel
year_dir = str(year)
if not os.path.exists(year_dir):
    os.mkdir(year_dir)
outputDir = os.path.join(year_dir,str(month)+"-"+str(day))
if not os.path.exists(outputDir):
    os.mkdir(outputDir)
timeStr = time.strftime('%Y-%m-%d_%H-%M-%S',time.localtime(time.time()))
writer = pd.ExcelWriter(outputDir+'/'+timeStr+'.xls')
sumSheet.to_excel(writer,'导入底稿',index=False)
writer.save()
print("\n\n================================")
print("报告已经输出在  ",outputDir," 目录下，文件名："+timeStr+".xls")

```
